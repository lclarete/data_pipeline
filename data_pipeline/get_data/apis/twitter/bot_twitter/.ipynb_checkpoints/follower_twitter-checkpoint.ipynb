## Connect to the api

```python
from twitter import Twitter, OAuth, TwitterHTTPError
```

```python
with open('config.txt', 'r') as f:
    config = f.readlines()
```

```python
OAUTH_TOKEN = config[0].split(':')[1][:-1]
OAUTH_SECRET = config[1].split(':')[1][:-1]
CONSUMER_KEY = config[2].split(':')[1][:-1]
CONSUMER_SECRET = config[3].split(':')[1][:-1]
```

```python
tw = Twitter(
    auth = (
    OAuth(
        OAUTH_TOKEN,
        OAUTH_SECRET,
        CONSUMER_KEY,
        CONSUMER_SECRET
        )
    )
)
```

# Build "beneficial profiles" database

## Search for profiles

* https://github.com/ideoforms/python-twitter-examples/blob/master/twitter-user-search.py

```python
cnn = tw.users.search(q='"cnn"')
```

```python
def profiles(obj):
    obj_id = []
    obj_screen_name = []
    for i in range(len(obj)):
        obj_id.append(obj[i]['id'])
        obj_screen_name.append(obj[i]['screen_name'])
    obj_users = dict(zip(obj_id, obj_screen_name))
    return(obj_users)
```

```python
cnn_users = profiles(cnn)
cnn_users
```

```python
president = tw.users.search(q='"president"')
```

```python
profiles(president)
```
