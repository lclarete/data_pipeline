```python
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)])
```

```python
a = text.apply(lemmatize_text)
```

# Data visualization

```python
from os import path
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def word_cloud_figure(col,
            max_words=None,
            background_color='white',
            stopwords=None):
    string_text = str(col)
    wordCloud = WordCloud(max_words=max_words,
                        repeat=False,
                        stopwords=stopwords,
                        background_color=background_color
                        ).generate(string_text)
    plt.figure(figsize=(10,7))
    plt.imshow(wordCloud, interpolation='bilinear')
    plt.axis('off')
    return(plt.show())
```

```python
sw_nltk = stopwords.words('english')
sw = sw_nltk + ['apartment', 'perfect', 'great', 'lovely', 'huge', 'location', 'well', 'place']
```

```python
word_cloud_figure(a, stopwords=sw)
```
