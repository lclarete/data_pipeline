```python
# drop duplicates
corpus = list(set(df.body.to_list()))
s = corpus[0]
s
```

# Preprocessing

This preprocessing will be used to extract:
* Working on the pronoms
* Punctuation;
* Drop some stopwords -- maintain the pronoms
* Lemmatization;

```python
# Working on the pronoms
# replace values

d = {
    "i’m": 'me',
    "i’ve": 'me',
    "i am": 'me',
    "i’ll": 'me',
    'myself': 'me',
    # it is trick to find a single letter, as i
    # using \s and replacing i with 'me' between spaces
    # https://stackoverflow.com/questions/7605198/how-does-b-work-when-using-regular-expressions
    '\si\s': ' me ',
    '\n':'',
    'she’s': 'she',
    'he’s': 'he',
    
    }

# lower cases
clean_corpus = list(map(lambda x: x.lower(), corpus))

# replacing words
clean_corpus = replacing_re_list(clean_corpus, d)
```

```python
# get the relation list
relation_list = get_relation_list(clean_corpus)

# get the entities list
entities_list = get_entities_list(clean_corpus)
```
